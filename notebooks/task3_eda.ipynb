## Create the Credit Risk Target
### Step 1: Create RFM Features

-  Using the original transaction dataset (df), not customer_df.

```python
import pandas as pd

snapshot_date = (
    pd.to_datetime(df["TransactionStartTime"]).max()
    + pd.Timedelta(days=1)
)

rfm = (
    df.groupby("CustomerId")
    .agg(
        Recency=(
            "TransactionStartTime",
            lambda x:
            (
                snapshot_date -
                pd.to_datetime(x).max()
            ).days
        ),

        Frequency=(
            "TransactionId",
            "count"
        ),

        Monetary=(
            "Amount",
            "sum"
        )
    )
)
```

## Step 2: Standardize RFM

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

rfm_scaled = scaler.fit_transform(rfm)
```

## Step 3: Cluster Customers

```python
from sklearn.cluster import KMeans

kmeans = KMeans(
    n_clusters=3,
    random_state=42
)

rfm["cluster"] = kmeans.fit_predict(
    rfm_scaled
)
```

## Step 4: Identify High-Risk Cluster

```python
rfm.groupby("cluster").mean()
```

```python
rfm["is_high_risk"] = (
    rfm["cluster"] == 2
).astype(int)
```

## Step 5: Merge with Customer Features

```python
customer_df = customer_df.merge(
    rfm[["is_high_risk"]],
    left_on="CustomerId",
    right_index=True
)
customer_df.head()
```

WoE transformation was explored using the xverse package. Due to compatibility limitations between the package and the current scikit-learn version, Information Value (IV) analysis was implemented directly to evaluate feature predictive strength. IV results identified the most informative customer behavioral features for credit risk prediction.

```python
# Simple IV Function
import pandas as pd
import numpy as np

def calculate_iv(df, feature, target):

    temp = pd.DataFrame()

    temp["feature"] = df[feature]
    temp["target"] = target

    temp["bin"] = pd.qcut(
        temp["feature"],
        q=10,
        duplicates="drop"
    )

    grouped = temp.groupby("bin")

    iv_table = pd.DataFrame()

    iv_table["good"] = grouped["target"].apply(
        lambda x: (x == 0).sum()
    )

    iv_table["bad"] = grouped["target"].apply(
        lambda x: (x == 1).sum()
    )

    iv_table["good_pct"] = (
        iv_table["good"] /
        iv_table["good"].sum()
    )

    iv_table["bad_pct"] = (
        iv_table["bad"] /
        iv_table["bad"].sum()
    )

    iv_table["woe"] = np.log(
        (iv_table["good_pct"] + 1e-6) /
        (iv_table["bad_pct"] + 1e-6)
    )

    iv_table["iv"] = (
        iv_table["good_pct"]
        - iv_table["bad_pct"]
    ) * iv_table["woe"]

    return iv_table["iv"].sum()
```

```python
# Compute IV
iv_scores = {}

numeric_cols = X.select_dtypes(
    include=["number"]
).columns

for col in numeric_cols:

    iv_scores[col] = calculate_iv(
        X,
        col,
        y
    )

iv_df = (
    pd.DataFrame(
        iv_scores.items(),
        columns=["Feature", "IV"]
    )
    .sort_values(
        "IV",
        ascending=False
    )
)

iv_df
```

# Standard IV Interpretation Scale
| IV Value    | Interpretation                                                     |
| ----------- | ------------------------------------------------------------------ |
| < 0.02      | Not predictive                                                     |
| 0.02 – 0.10 | Weak predictor                                                     |
| 0.10 – 0.30 | Medium predictor                                                   |
| 0.30 – 0.50 | Strong predictor                                                   |
| > 0.50      | Very strong predictor                                              |
| > 1.00      | Often indicates target leakage or an issue requiring investigation |

### Key Findings
#### 1. avg_transaction_month (IV = 8.709)

-  This is extraordinarily high.

IV = 8.709

In credit risk modeling, IV values above 1.0 are uncommon and often suggest:

-  Target leakage
-  Artificial relationship with the target
-  Features used indirectly during target creation

Since our target (is_high_risk) was generated from transaction history and recency-related behavior, the transaction month may be strongly linked to how customers were assigned to clusters.

### Interpretation for the report:
Average transaction month exhibited extremely high predictive power. However, the unusually large IV value suggests potential target leakage or a strong dependency between the proxy target creation process and transaction timing variables. This feature should be investigated carefully before deployment.

# Report Summary
Information Value analysis revealed that transaction behavior features are highly predictive of customer risk. Transaction count, total transaction amount, and transaction amount variability emerged as the most informative behavioral predictors. Temporal variables such as average transaction month and year produced exceptionally high IV scores, suggesting possible target leakage due to their close relationship with the RFM-based proxy target construction process. Overall, customer activity frequency and monetary behavior were found to be the strongest indicators of credit risk, supporting their inclusion in subsequent model development.
